```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('data_2/track_features/tf_mini.csv')
```

```python
list(df.columns)
```

```python
fig = px.histogram(df, x='beat', nbins=10)
fig.show()
```

```python
df_grouped = df.groupby('release_year')['track_id'].count().rename('Count').to_frame().reset_index()
df_grouped
```

```python
fig = px.line(df_grouped, x='release_year', y='Count')
fig.show()
```
